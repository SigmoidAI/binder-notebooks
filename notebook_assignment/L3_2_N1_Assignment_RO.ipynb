# Tema de programare: Detectarea si eliminarea duplicatelor


Bun venit la tema despre **Detectarea si eliminarea duplicatelor**!

Inregistrarile duplicate sunt o problema comuna de calitate a datelor. Ele apar din erori de introducere a datelor, surse de date combinate, trimiteri repetate ale formularelor si migrari de sistem. Lasate netratate, duplicatele distorsioneaza statisticile, introduc bias in antrenarea modelelor si umfla metricile de evaluare.

Exista doua tipuri principale de duplicate:

| Tip | Descriere | Exemplu |
|------|-------------|---------|
| **Duplicate exacte** | Fiecare camp este identic | Acelasi rand copiat de doua ori |
| **Aproape duplicate** | Majoritatea campurilor coincid; exista o mica variatie intr-un camp | "Alice Smith" vs "ALICE SMITH" |

**Cand sa elimini duplicatele:**
- Inainte de antrenarea modelului - duplicatele din setul de antrenare pot provoca scurgere de date in setul de test.
- Inainte de agregari - numararile, mediile si sumele sunt umflate de randurile repetate.
- Inainte de imbinarea cu date externe - cheile duplicate provoaca multiplicarea randurilor.

**Ce vei face in aceasta tema:**

* Vei detecta randurile duplicate exacte folosind metodele built-in din pandas
* Vei elimina duplicatele exacte pastrand ordinea originala a randurilor
* Vei identifica siruri aproape duplicate cu un scor de fuzzy matching
* Vei marca randurile duplicate cu o coloana booleana pentru utilizare ulterioara
* Vei produce un raport structurat de deduplicare cu metrici-cheie

## 4 - Detectarea fuzzy / a aproape duplicatelor

Potrivirea exacta rateaza inregistrarile care sunt **aproape identice**, dar difera prin capitalizare, spatiere sau mici greseli de tiparire. **Fuzzy matching** calculeaza un scor de similaritate intre doua siruri si marcheaza perechile peste un prag ales.

Modulul din biblioteca standard Python `difflib.SequenceMatcher` ofera o metoda `ratio()` care returneaza o valoare in $[0, 1]$ ce reprezinta cat de similare sunt doua siruri. Noi o scalăm la $[0, 100]$ pentru lizibilitate.
