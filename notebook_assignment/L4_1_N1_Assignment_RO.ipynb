# Tema de Programare: Fundamente Matplotlib


Bine ai venit la tema despre **Fundamente Matplotlib**!

Matplotlib este cea mai populară bibliotecă de vizualizare din Python. La bază,
folosește o **interfață orientată pe obiecte (OO)** construită în jurul a două obiecte-cheie:

| Obiect | Descriere |
|--------|-----------|
| **Figure** | Containerul de nivel superior — ca o pânză de desen |
| **Axes** | O singură zonă de grafic din Figure (pot exista mai multe într-o Figure) |
| **Artists** | Tot ceea ce este desenat: linii, text, patch-uri, tick-uri etc. |

> **Pont:** Preferă întotdeauna interfața OO (`fig, ax = plt.subplots()`) în locul
> apelurilor simple `plt.xxx()` — îți oferă control explicit asupra fiecărui obiect.

## 7 - Punem Totul Cap la Cap

Felicitări pentru finalizarea tuturor celor 5 exerciții!

Celula de mai jos combină toate funcțiile tale pentru a produce un **dashboard de vânzări 2×2** complet:
1. `create_subplots` — construiește layout-ul 2×2
2. `add_line_chart` — desenează trendul în timp

````python
# Dashboard complet de vânzări folosind toate cele 5 funcții
fig, axes = create_subplots(nrows=2, ncols=2, figsize=(14, 9))

sorted_df = df.sort_values('order_date').reset_index(drop=True)

# ── Stânga sus: trendul vânzărilor cu adnotarea vârfului ─────────────────────
add_line_chart(
    axes[0, 0], sorted_df['order_date'], sorted_df['amount'],
    title='Vânzări în Timp', xlabel='Dată', ylabel='Valoare (USD)',
)
add_peak_annotation(axes[0, 0], sorted_df, x_column='order_date', y_column='amount')
apply_clean_styling(axes[0, 0])

# ── Dreapta sus: valoarea medie pe regiune ───────────────────────────────────
region_means = df.groupby('region')['amount'].mean()
axes[0, 1].bar(region_means.index, region_means.values,
               color='#388E3C', alpha=0.85, edgecolor='white')
axes[0, 1].set_title('Valoare Medie pe Regiune')
axes[0, 1].set_xlabel('Regiune')
axes[0, 1].set_ylabel('Valoare Medie (USD)')
apply_clean_styling(axes[0, 1])

# ── Stânga jos: histogramă pentru distribuția valorilor ──────────────────────
axes[1, 0].hist(df['amount'], bins=15, color='#F57C00', alpha=0.8, edgecolor='white')
axes[1, 0].set_title('Distribuția Valorilor')
axes[1, 0].set_xlabel('Valoare (USD)')
axes[1, 0].set_ylabel('Număr')
apply_clean_styling(axes[1, 0])

# ── Dreapta jos: scatter unități vs valoare ──────────────────────────────────
axes[1, 1].scatter(df['units'], df['amount'],
                   alpha=0.6, color='#6A1B9A', edgecolors='white', linewidths=0.5)
axes[1, 1].set_title('Unități vs Valoare')
axes[1, 1].set_xlabel('Unități')
axes[1, 1].set_ylabel('Valoare (USD)')
apply_clean_styling(axes[1, 1])

plt.suptitle('Dashboard de Vânzări', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
save_figure(fig, '/tmp/L4_1_N1_dashboard.png', dpi=150)
print('Dashboard salvat în /tmp/L4_1_N1_dashboard.png')
plt.show()
````
