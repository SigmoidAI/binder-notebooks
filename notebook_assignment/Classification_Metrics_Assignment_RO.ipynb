## 1 - Introducere in metricile de clasificare

**Metricile de clasificare** evalueaza cat de bine distinge un model intre clase diferite. Spre deosebire de metricile de regresie, care masoara magnitudinea erorii de predictie, metricile de clasificare se concentreaza pe corectitudinea atribuirii claselor.

### Fundamentele clasificarii binare:

Pentru clasificarea binara (pozitiv/negativ), predictiile se incadreaza in patru categorii:

- **True Positives (TP)**: instante pozitive prezise corect
- **True Negatives (TN)**: instante negative prezise corect
- **False Positives (FP)**: instante negative prezise incorect ca pozitive (eroare de tip I)
- **False Negatives (FN)**: instante pozitive prezise incorect ca negative (eroare de tip II)

### Metrici cheie:

**Accuracy**: corectitudine generala
$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

**Precision**: calitatea predictiilor pozitive
$$\text{Precision} = \frac{TP}{TP + FP}$$

**Recall (Sensitivity, TPR)**: acoperirea exemplelor cu adevarat pozitive
$$\text{Recall} = \frac{TP}{TP + FN}$$

**F1-Score**: media armonica dintre precision si recall
$$F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$$

**Specificity (TNR)**: acoperirea exemplelor cu adevarat negative
$$\text{Specificity} = \frac{TN}{TN + FP}$$

### Selectarea metricii:

- **Set de date echilibrat**: accuracy functioneaza bine
- **Set de date dezechilibrat**: foloseste precision, recall, F1 sau AUC
- **Sensibil la costuri**: optimizeaza in functie de costurile FP/FN
- **Diagnostic medical**: prioritizeaza recall-ul (evita ratarea cazurilor pozitive)
- **Detectie spam**: prioritizeaza precision-ul (evita alarmele false)

### Pragul de clasificare:

Majoritatea clasificatorilor returneaza probabilitati $P(y=1|x)$. Pragul $\tau$ determina clasificarea:
$$\hat{y} = \begin{cases} 1 & \text{if } P(y=1|x) \geq \tau \\ 0 & \text{otherwise} \end{cases}$$

Valoarea implicita este $\tau = 0.5$, dar poate fi optimizata in functie de cerinte.

### Functii ajutatoare


````python
def plot_confusion_matrix(cm, title="Confusion Matrix", labels=None, normalize=False):
    """Plot confusion matrix with annotations."""
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        fmt = '.2f'
    else:
        fmt = 'd'
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt=fmt, cmap='Blues', 
                xticklabels=labels, yticklabels=labels,
                cbar_kws={'label': 'Count' if not normalize else 'Proportion'})
    plt.ylabel('Actual', fontsize=12)
    plt.xlabel('Predicted', fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()


def plot_metrics_vs_threshold(y_true, y_pred_proba, metrics=['precision', 'recall']):
    """Plot how metrics change with classification threshold."""
    thresholds = np.linspace(0, 1, 100)
    metric_values = {metric: [] for metric in metrics}
    
    for threshold in thresholds:
        y_pred = (y_pred_proba >= threshold).astype(int)
        
        if 'precision' in metrics:
            metric_values['precision'].append(
                precision_score(y_true, y_pred, zero_division=0)
            )
        if 'recall' in metrics:
            metric_values['recall'].append(
                recall_score(y_true, y_pred, zero_division=0)
            )
        if 'f1' in metrics:
            metric_values['f1'].append(
                f1_score(y_true, y_pred, zero_division=0)
            )
    
    plt.figure(figsize=(10, 6))
    for metric, values in metric_values.items():
        plt.plot(thresholds, values, linewidth=2, label=metric.capitalize())
    
    plt.xlabel('Threshold', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.title('Metrics vs Classification Threshold', fontsize=14, fontweight='bold')
    plt.legend(loc='best')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_roc_curve(y_true, y_pred_proba, label=None):
    """Plot ROC curve with AUC score."""
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
    auc = roc_auc_score(y_true, y_pred_proba)
    
    plt.figure(figsize=(10, 8))
    plt.plot(fpr, tpr, linewidth=2, label=f'{label} (AUC = {auc:.3f})' if label else f'AUC = {auc:.3f}')
    plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Random Classifier')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curve', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return auc


def plot_pr_curve(y_true, y_pred_proba, label=None):
    """Plot Precision-Recall curve."""
    precision, recall, _ = precision_recall_curve(y_true, y_pred_proba)
    ap = average_precision_score(y_true, y_pred_proba)
    
    plt.figure(figsize=(10, 8))
    plt.plot(recall, precision, linewidth=2, 
             label=f'{label} (AP = {ap:.3f})' if label else f'AP = {ap:.3f}')
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
    plt.legend(loc='best')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return ap


print("Helper functions defined!")
````

## 2 - Analiza matricei de confuzie

**Matricea de confuzie** este un instrument fundamental pentru intelegerea performantei clasificarii. Ea arata numarul de true positives, true negatives, false positives si false negatives.

### Structura:

```
                Prezis Negativ    Prezis Pozitiv
Actual Negativ        TN                FP
Actual Pozitiv        FN                TP
```

### Metrici derivate:

Din matricea de confuzie putem calcula:

- **True Positive Rate (TPR/Recall/Sensitivity)**: $\frac{TP}{TP + FN}$
- **True Negative Rate (TNR/Specificity)**: $\frac{TN}{TN + FP}$
- **False Positive Rate (FPR)**: $\frac{FP}{FP + TN} = 1 - \text{Specificity}$
- **False Negative Rate (FNR)**: $\frac{FN}{FN + TP} = 1 - \text{Recall}$

### Interpretare:

- **Elementele de pe diagonala** (TN, TP): predictii corecte
- **Elementele din afara diagonalei** (FP, FN): erori
- **Normalizare pe randuri**: arata proportia fiecarei clase reale
- **Normalizare pe coloane**: arata precision pentru fiecare clasa prezisa

## 3 - Precision, Recall si F1 Score

**Precision** si **Recall** ofera perspective complementare asupra performantei clasificarii:

### Precision:
$$\text{Precision} = \frac{TP}{TP + FP}$$

- Raspunde la intrebarea: "Din toate predictiile pozitive, cate au fost corecte?"
- Precision mare -> putine alarme false
- Important cand false positive sunt costisitoare

### Recall:
$$\text{Recall} = \frac{TP}{TP + FN}$$

- Raspunde la intrebarea: "Din toate cazurile realmente pozitive, cate am gasit?"
- Recall mare -> putine cazuri ratate
- Important cand false negative sunt costisitoare

### Compromisul precision-recall:

- **Prag mai mic** -> Mai multe predictii pozitive -> Recall mai mare, precision mai mic
- **Prag mai mare** -> Mai putine predictii pozitive -> Recall mai mic, precision mai mare
- De obicei nu le poti maximiza simultan pe ambele

### F1-Score:
$$F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} = \frac{2 \cdot TP}{2 \cdot TP + FP + FN}$$

- Media armonica dintre precision si recall
- Echilibreaza ambele metrici
- Penalizeaza dezechilibrul extrem dintre precision si recall
- Util atunci cand ai nevoie de o singura metrica

### F-Beta Score:
$$F_\beta = (1 + \beta^2) \cdot \frac{\text{Precision} \cdot \text{Recall}}{\beta^2 \cdot \text{Precision} + \text{Recall}}$$

- $\beta < 1$: pune accent pe precision
- $\beta = 1$: echilibrat (F1-score)
- $\beta > 1$: pune accent pe recall

## 4 - Curbe ROC si PR

Curbele **ROC (Receiver Operating Characteristic)** si **PR (Precision-Recall)** vizualizeaza performanta modelului pe toate pragurile de clasificare.

### Curba ROC:

Reprezinta TPR (Recall) vs FPR la praguri diferite:
- **TPR**: $\frac{TP}{TP + FN}$ (axa y)
- **FPR**: $\frac{FP}{FP + TN}$ (axa x)

**Interpretare:**
- Linia diagonala: clasificator aleator (AUC = 0.5)
- Coltul din stanga sus: clasificator perfect (AUC = 1.0)
- Area Under Curve (AUC): rezumat intr-un singur numar
- AUC = probabilitatea ca modelul sa evalueze un pozitiv aleator mai sus decat un negativ aleator

### Curba PR:

Reprezinta Precision vs Recall la praguri diferite:
- **Recall**: $\frac{TP}{TP + FN}$ (axa x)
- **Precision**: $\frac{TP}{TP + FP}$ (axa y)

**Interpretare:**
- Coltul din dreapta sus: clasificator perfect
- Baza: proportia clasei pozitive
- Average Precision (AP): aria de sub curba PR
- Mai informativa decat ROC pentru seturi de date dezechilibrate

### Cand folosesti fiecare:

**Curba ROC:**
- Seturi de date echilibrate
- Iti pasa in mod egal de ambele clase
- Compari mai multe modele

**Curba PR:**
- Seturi de date dezechilibrate
- Clasa pozitiva este rara si importanta
- Accent pe performanta clasei pozitive

## 5 - Optimizarea pragului

Pragul implicit de clasificare de 0.5 este adesea suboptimal. Pragul optim depinde de:

### Obiective de business:

**Diagnostic medical:**
- False Negative (boala ratata): foarte costisitor
- False Positive (teste inutile): mai putin costisitor
- **Solutie**: prag mai mic, prioritizeaza recall

**Detectie spam:**
- False Positive (blocarea unui email important): foarte costisitor
- False Negative (spam in inbox): mai putin costisitor
- **Solutie**: prag mai mare, prioritizeaza precision

### Clasificare sensibila la costuri:

Defineste functia de cost:
$$\text{Cost} = C_{FP} \times FP + C_{FN} \times FN$$

Gaseste pragul care minimizeaza costul total:
$$\tau^* = \arg\min_{\tau} \text{Cost}(\tau)$$

### Metode de optimizare:

1. **Minimizarea costului**: minimizeaza suma ponderata a erorilor
2. **Maximizarea F-Beta**: maximizeaza F-score cu $\beta$ personalizat
3. **Indicele lui Youden**: maximizeaza $J = \text{TPR} - \text{FPR}$
4. **Metrica tinta**: atinge un anumit nivel de recall/precision

### Consideratii practice:

- Valideaza pragul pe un set de date separat
- Monitorizeaza performanta in timp
- Ia in calcul incertitudinea estimarilor de cost
- Documenteaza rationamentul deciziei

## 6 - Metrici Multi-Clasa

Extinderea metricilor de clasificare binara la mai multe clase necesita strategii de agregare.

### Metode de agregare:

**Media macro:**
$$\text{Macro-Precision} = \frac{1}{K} \sum_{k=1}^{K} \text{Precision}_k$$

- Calculeaza metrica pentru fiecare clasa independent
- Face media neponderata
- Trateaza toate clasele la fel
- Buna pentru seturi de date echilibrate
- Sensibila la performanta claselor minoritare

**Media micro:**
$$\text{Micro-Precision} = \frac{\sum_{k=1}^{K} TP_k}{\sum_{k=1}^{K} (TP_k + FP_k)}$$

- Agrega contributiile tuturor claselor
- Calculeaza metrica din TP, FP, FN agregate
- Fiecare esantion contribuie egal
- Buna pentru seturi de date dezechilibrate
- Micro-F1 este egal cu accuracy in multi-clasa

**Media ponderata:**
$$\text{Weighted-Precision} = \sum_{k=1}^{K} w_k \cdot \text{Precision}_k$$

unde $w_k = \frac{n_k}{n}$ (proportia clasei k)

- Pondereaza fiecare clasa dupa frecventa ei
- Tine cont de dezechilibrul claselor
- Cea mai folosita in practica

### Matricea de confuzie multi-clasa:

- Matrice $K \times K$ pentru K clase
- Diagonala: predictii corecte
- In afara diagonalei: confuzii intre perechi specifice de clase
- Utila pentru a identifica ce clase sunt confundate

### One-vs-Rest (OvR):

Pentru fiecare clasa k:
- Pozitiv: clasa k
- Negativ: toate celelalte clase
- Calculeaza metrici binare
- Agrega folosind macro/micro/weighted

$$F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} = \frac{2 \cdot TP}{2 \cdot TP + FP + FN}$$

### Idei principale:

- **Recall**: cand false negative sunt costisitoare (medical, securitate)
- **F1**: echilibru intre precision si recall
- **AUC-ROC**: capacitatea generala de discriminare a modelului
- **AUC-PR**: mai buna pentru seturi de date dezechilibrate

**Bune practici:**
- Examineaza intotdeauna mai intai matricea de confuzie

- Valideaza pe un set de test separat
- Documenteaza alegerile de metrici si rationalele

**Capcane frecvente:**

1. **Sa te bazezi doar pe accuracy** pentru seturi de date dezechilibrate
   - Solutie: foloseste precision, recall, F1 sau AUC

2. **Sa ignori pragul de clasificare**
   - Solutie: optimizeaza pragul in functie de costuri

3. **Sa folosesti metrica gresita pentru problema**
   - Solutie: aliniaza metrica la cerintele de business

4. **Sa supra-optimizezi dupa metricile de validare**
   - Solutie: foloseste un set de test separat pentru evaluarea finala

5. **Sa compari modele cu praguri diferite**
   - Solutie: foloseste metrici bazate pe probabilitati (AUC)

**Diagnostic medical:**
- Recall mare pentru a evita omiterea bolilor

- Echilibrezi precision si recall
- Minimizezi false positive (frustrarea clientilor)
- Minimizezi false negative (pierdere financiara)

- Precision at K pentru motoare de cautare
