# Tema de Programare: Analiză Detaliată a Variabilei Țintă


## Bine ai venit

În machine learning-ul supervizat, **variabila țintă** este coloana pe care încerci să o prezici. Înainte de a antrena orice model, trebuie să-i înțelegi structura în profunzime — pentru că problemele din țintă îți corup întregul pipeline.

### De ce contează

**Dezechilibrul de clasă** este una dintre cele mai frecvente capcane în clasificarea binară. Când clasa minoritară reprezintă doar 18 % din observații (ca în coloana noastră `returned`), un model naiv poate obține 82 % acuratețe doar *prezicând mereu clasa majoritară* — și totuși să fie complet inutil la detectarea retururilor.

> **Pont:** Pentru ținte dezechilibrate, folosește întotdeauna precision / recall / F1, niciodată doar accuracy.

**Anomaliile din variabila țintă** sunt la fel de periculoase. Valori neașteptate (de ex. `2` într-o coloană binară) sau valori nule în coloana țintă îți vor distorsiona în tăcere modelul în timpul antrenării dacă nu le detectezi din timp.

### Ce vei realiza

În acest notebook vei:
1. Calcula un tabel de echilibru al claselor și cuantifica dezechilibrul.
2. Clasifica severitatea dezechilibrului folosind o scală pe patru niveluri.
3. Construi un grafic cu bare de calitate pentru publicare al distribuției țintei.
4. Calcula mediile caracteristicilor pe clasă pentru a evidenția asocierile caracteristică–țintă.
5. Construi un detector robust de anomalii pentru coloanele țintă.

**Set de date:** 300 de comenzi e-commerce simulate. Țintă: `returned` (0 = păstrat, 1 = returnat).

<a name='7---putting-it-all-together'></a>
## 7 - Punem Totul Cap la Cap

Acum combinăm toate cele cinci funcții într-un raport complet de analiză a variabilei țintă.


```python
# Pipeline complet pentru analiza țintei
TARGET = 'returned'
NUMERIC_COLS = ['order_value', 'customer_age', 'days_to_ship', 'items_ordered', 'days_since_last_purchase']

balance = get_class_balance(df, TARGET)
minority_prop = balance['proportion'].min()
severity = classify_imbalance(minority_prop)
stats = compute_target_stats(df, TARGET, NUMERIC_COLS)
anomaly_report = detect_target_anomalies(df, TARGET)

print("=" * 60)
print(f"  ANALIZA VARIABILEI ȚINTĂ  —  coloana: '{TARGET}'")
print("=" * 60)
print("\n[ Echilibrul Claselor ]")
print(balance.to_string())
print(f"\n[ Severitatea Dezechilibrului ]")
helper_utils.print_imbalance_report(minority_prop)
print(f"\n[ Verificarea Anomaliilor ]")
for k, v in anomaly_report.items():
    print(f"  {k}: {v}")
print(f"\n[ Mediile Caracteristicilor pe Clasă ]")
print(stats.to_string())
```

```python
# Vizualizare alăturată
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(14, 5))

# Stânga: grafic cu bare pentru distribuția claselor
counts = df[TARGET].value_counts().sort_index()
proportions = counts / len(df)
labels = [str(v) for v in counts.index]
bars = ax_left.bar(labels, counts.values, color=['#4C72B0', '#C44E52'], edgecolor='white')
for bar, prop in zip(bars, proportions.values):
    h = bar.get_height()
    ax_left.text(
        bar.get_x() + bar.get_width() / 2, h + 1,
        f"{prop:.1%}", ha='center', va='bottom', fontsize=11, fontweight='bold',
    )
ax_left.set_title(f"{TARGET} — Distribuția Claselor", fontsize=13)
ax_left.set_xlabel(TARGET)
ax_left.set_ylabel("Număr")

# Dreapta: bară orizontală — media order_value și customer_age pe clasă
feature_stats = compute_target_stats(df, TARGET, ['order_value', 'customer_age'])
x = np.arange(len(feature_stats.columns))
width = 0.3
colors_cls = ['#4C72B0', '#C44E52']
for i, (cls_val, row) in enumerate(feature_stats.iterrows()):
    ax_right.barh(
        x + i * width, row.values, height=width,
        label=f"{TARGET}={cls_val}", color=colors_cls[i], edgecolor='white',
    )
ax_right.set_yticks(x + width / 2)
ax_right.set_yticklabels(feature_stats.columns)
ax_right.set_xlabel("Valoare medie")
ax_right.set_title("Valorile Medii ale Caracteristicilor după Clasa Țintă", fontsize=13)
ax_right.legend()

plt.tight_layout()
plt.show()
```
