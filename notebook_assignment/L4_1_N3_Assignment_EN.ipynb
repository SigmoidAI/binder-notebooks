# Programming Assignment: Interactive Visualization with Plotly Express


Welcome to the **Interactive Visualization with Plotly Express** assignment!

Plotly's key advantage over static libraries is **interactivity** — every chart you create supports hover tooltips, zoom, pan, and a one-click PNG download out of the box. No extra configuration needed.

Plotly exposes two levels of API:

| API Level | Description | Returns |
|-----------|-------------|--------|
| **`plotly.express` (px)** | High-level, one-liner charts | `go.Figure` |
| **`plotly.graph_objects` (go)** | Low-level, full control | `go.Figure` |

> **Tip:** Every `px.*` function returns a `go.Figure`. Use `fig.update_layout()` and `fig.update_traces()` to refine any chart created by either API.

**What You Will Do in This Assignment:**

* Create interactive scatter plots with colour grouping and hover information
* Build time-series line charts grouped by category
* Construct grouped bar charts to compare categories
* Customise figure layouts with titles, axis labels, and font settings
* Export fully interactive charts as self-contained HTML files

## 7 - Putting It All Together

Now let's combine all five functions to build a mini sales dashboard. We will display three coordinated views of the same dataset:

1. **Scatter** — revenue vs units, coloured by category
2. **Time-series** — revenue evolution, coloured by category
3. **Grouped bar** — revenue per category, sub-grouped by delivery status

Each chart is fully customised and the final scatter view is exported as an HTML file.


```python
# --- Chart 1: Scatter ---
scatter = create_scatter(df, x='units', y='amount', color='category',
                         hover_data=['region', 'status'])
scatter = customize_layout(scatter, title='Revenue vs Units (by Category)',
                           xaxis_title='Units Ordered',
                           yaxis_title='Revenue (USD)')
scatter.show()

# --- Chart 2: Time-series ---
ts = create_time_series(df, date_col='order_date', value_col='amount',
                        color_col='category')
ts = customize_layout(ts, title='Weekly Revenue by Category',
                      xaxis_title='Order Date',
                      yaxis_title='Revenue (USD)')
ts.show()

# --- Chart 3: Grouped bar ---
bar = create_grouped_bar(df, x='category', y='amount', color='status')
bar = customize_layout(bar, title='Revenue by Category & Delivery Status',
                       xaxis_title='Category',
                       yaxis_title='Revenue (USD)')
bar.show()

# --- Export the scatter chart ---
save_figure_html(scatter, '/tmp/L4_1_N3_dashboard_scatter.html')
print('Dashboard scatter saved to /tmp/L4_1_N3_dashboard_scatter.html')
```
