# Programming Assignment: Matplotlib Fundamentals


Welcome to the **Matplotlib Fundamentals** assignment!

Matplotlib is Python's most popular plotting library. At its core, it uses an
**object-oriented (OO) interface** built around two key objects:

| Object | Description |
|--------|-------------|
| **Figure** | The top-level container — like a canvas |
| **Axes** | A single plot area within the Figure (can have many per Figure) |
| **Artists** | Everything drawn: lines, text, patches, ticks, etc. |

> **Tip:** Always prefer the OO interface (`fig, ax = plt.subplots()`) over
> bare `plt.xxx()` calls — it gives you explicit control over every object.

## 7 - Putting It All Together

Congratulations on completing all 5 exercises!

The cell below combines all your functions to produce a complete **2×2 sales dashboard**:
1. `create_subplots` — builds the 2×2 layout
2. `add_line_chart` — draws the time-series trend
3. `add_peak_annotation` — marks the highest-value order
4. `apply_clean_styling` — cleans up all four panels
5. `save_figure` — exports the dashboard to disk

````python
# Full sales dashboard using all 5 functions
fig, axes = create_subplots(nrows=2, ncols=2, figsize=(14, 9))

sorted_df = df.sort_values('order_date').reset_index(drop=True)

# ── Top-left: sales trend with peak annotation ───────────────────────────────
add_line_chart(
    axes[0, 0], sorted_df['order_date'], sorted_df['amount'],
    title='Sales Over Time', xlabel='Date', ylabel='Amount (USD)',
)
add_peak_annotation(axes[0, 0], sorted_df, x_column='order_date', y_column='amount')
apply_clean_styling(axes[0, 0])

# ── Top-right: average amount by region ─────────────────────────────────────
region_means = df.groupby('region')['amount'].mean()
axes[0, 1].bar(region_means.index, region_means.values,
               color='#388E3C', alpha=0.85, edgecolor='white')
axes[0, 1].set_title('Avg Amount by Region')
axes[0, 1].set_xlabel('Region')
axes[0, 1].set_ylabel('Avg Amount (USD)')
apply_clean_styling(axes[0, 1])

# ── Bottom-left: amount distribution histogram ───────────────────────────────
axes[1, 0].hist(df['amount'], bins=15, color='#F57C00', alpha=0.8, edgecolor='white')
axes[1, 0].set_title('Amount Distribution')
axes[1, 0].set_xlabel('Amount (USD)')
axes[1, 0].set_ylabel('Count')
apply_clean_styling(axes[1, 0])

# ── Bottom-right: units vs amount scatter ───────────────────────────────────
axes[1, 1].scatter(df['units'], df['amount'],
                   alpha=0.6, color='#6A1B9A', edgecolors='white', linewidths=0.5)
axes[1, 1].set_title('Units vs Amount')
axes[1, 1].set_xlabel('Units')
axes[1, 1].set_ylabel('Amount (USD)')
apply_clean_styling(axes[1, 1])

plt.suptitle('Sales Dashboard', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
save_figure(fig, '/tmp/L4_1_N1_dashboard.png', dpi=150)
print('Dashboard saved to /tmp/L4_1_N1_dashboard.png')
plt.show()
````
