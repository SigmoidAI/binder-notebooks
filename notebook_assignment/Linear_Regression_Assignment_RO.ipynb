Regresia liniara simpla modeleaza relatia dintre o singura trasatura $x$ si o variabila tinta $y$ folosind ecuatia:

$$h(x) = \theta_0 + \theta_1 x$$

unde $\theta_0$ este interceptul, iar $\theta_1$ reprezinta panta dreptei. Obiectivul este sa gasim parametrii care minimizeaza eroarea dintre predictii si valorile reale.

```python
def gradient_descent(X, theta, y, learning_rate, iterations):
    """
    Perform gradient descent to learn theta.

    Args:
        X: Feature array of shape (m, n)
        theta: Initial parameters of shape (n,)
        y: Target array of shape (m,)
        learning_rate: Learning rate
        iterations: Number of iterations

    Returns:
        theta: Optimized parameters
        cost_history: List of cost values over iterations
    """
    m = len(y)
    cost_history = []
    
    for i in range(iterations):
        ### ÎNCEPUT DE COD AICI ### (≈ 4 lines)
        y_pred = None
        gradient = None
        theta = None
        
        # Record cost
        cost = None
        ### SFÂRȘIT DE COD AICI ###
        
        cost_history.append(cost)
        
        if i % 100 == 0:
            print(f"Iteration {i:4d}: Cost = {cost:.4f}")
    
    return theta, cost_history
```
