# Programming Assignment: Data Type Conversion

Welcome to the **Data Type Conversion** assignment!

In real-world datasets, data is often stored or imported with incorrect types — numbers stored as strings, dates as plain text, or categories as unordered objects. Fixing these types is not just cosmetic:

| Reason | Why It Matters |
|--------|----------------|
| **Memory efficiency** | `category` and `int` types use far less memory than `object` |
| **Correct operations** | Arithmetic on strings silently fails or produces wrong results |
| **Downstream models** | Scikit-learn estimators require numeric input; datetime features unlock time-based signals |
