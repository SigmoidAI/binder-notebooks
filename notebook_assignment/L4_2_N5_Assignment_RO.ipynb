Valorile variază de la **-1** (negativ perfect) la **+1** (pozitiv perfect),
iar **0** înseamnă lipsa unei relații liniare.

<a name='7---putting-it-all-together'></a>
## 7 - Punem Totul Cap la Cap

Excelent! Hai să combinăm toate cele cinci analize într-un singur dashboard ca să obținem
imaginea completă a relațiilor dintre caracteristici și țintă dintr-o singură privire.


```python
import matplotlib.gridspec as gridspec

numeric_cols = [
    "order_value", "days_to_ship", "items_ordered",
    "customer_age", "days_since_last_purchase"
]
target_col = "returned"

# Construiește toate rezumatele
summary     = build_relationship_summary(df, numeric_cols, "region", target_col)
correlations = compute_feature_target_correlations(df, numeric_cols, target_col)
group_means  = compute_group_means(df, "region", target_col)

fig = plt.figure(figsize=(18, 10))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.50, wspace=0.38)

# ── Stânga sus: grafic cu bare pentru corelații ──────────────────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
sorted_corr  = correlations.sort_values()
colors_corr  = ['#4C72B0' if v >= 0 else '#C44E52' for v in sorted_corr.values]
ax1.barh(sorted_corr.index, sorted_corr.values, color=colors_corr, edgecolor='white')
ax1.axvline(0, color='black', linewidth=0.8)
ax1.set_title(f"Corelațiile Caracteristicilor cu {target_col}", fontsize=10)
ax1.set_xlabel("Corelație Pearson", fontsize=9)

# ── Sus centru: KDE pentru caracteristica de top ────────────────────────────────────
top_feature = summary.index[0]
ax2 = fig.add_subplot(gs[0, 1])
classes    = sorted(df[target_col].unique())
colors_kde = ['#4C72B0', '#C44E52']
for i, cls in enumerate(classes):
    subset = df.loc[df[target_col] == cls, top_feature].dropna()
    subset.plot.kde(ax=ax2, color=colors_kde[i], label=f"{target_col}={cls}")
    ax2.axvline(subset.mean(), color=colors_kde[i], linestyle='--', linewidth=1.2, alpha=0.8)
ax2.set_title(f"{top_feature} KDE după {target_col}", fontsize=10)
ax2.set_xlabel(top_feature, fontsize=9)
ax2.legend(fontsize=8)

# ── Sus dreapta: KDE pentru a doua caracteristică ───────────────────────────────────
second_feature = summary.index[1] if len(summary) > 1 else numeric_cols[1]
ax3 = fig.add_subplot(gs[0, 2])
for i, cls in enumerate(classes):
    subset = df.loc[df[target_col] == cls, second_feature].dropna()
    subset.plot.kde(ax=ax3, color=colors_kde[i], label=f"{target_col}={cls}")
    ax3.axvline(subset.mean(), color=colors_kde[i], linestyle='--', linewidth=1.2, alpha=0.8)
ax3.set_title(f"{second_feature} KDE după {target_col}", fontsize=10)
ax3.set_xlabel(second_feature, fontsize=9)
ax3.legend(fontsize=8)

# ── Jos stânga: bar chart pentru mediile pe grupuri ─────────────────────────────────
ax4 = fig.add_subplot(gs[1, 0])
bar_colors = ['#C44E52' if v == group_means.max() else '#4C72B0' for v in group_means.values]
ax4.bar(group_means.index, group_means.values, color=bar_colors, edgecolor='white')
ax4.set_title("Rata de Retur pe Regiune", fontsize=10)
ax4.set_xlabel("Regiune", fontsize=9)
ax4.set_ylabel("Rata Medie de Retur", fontsize=9)
ax4.tick_params(axis='x', labelsize=8)

# ── Jos dreapta: tabel rezumat al relațiilor ────────────────────────────────────────
ax5 = fig.add_subplot(gs[1, 1:])
ax5.axis('off')
col_labels = ['Caracteristică', 'Corelație cu Ținta', 'Media (0)', 'Media (1)', 'Diferență Medii']
table_data = [
    [idx,
     f"{row['correlation_with_target']:.4f}",
     f"{row['mean_class_0']:.2f}",
     f"{row['mean_class_1']:.2f}",
     f"{row['mean_diff']:.2f}"]
    for idx, row in summary.iterrows()
]
tbl = ax5.table(cellText=table_data, colLabels=col_labels,
                cellLoc='center', loc='center', bbox=[0, 0.05, 1, 0.9])
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
ax5.set_title("Tabel Rezumat al Relațiilor", fontsize=10, pad=12)

plt.suptitle("Dashboard al Relațiilor dintre Caracteristici și Țintă",
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
numeric_cols = [
    "order_value", "days_to_ship", "items_ordered",
    "customer_age", "days_since_last_purchase"
]
summary      = build_relationship_summary(df, numeric_cols, "region", "returned")
correlations = compute_feature_target_correlations(df, numeric_cols, "returned")
group_means  = compute_group_means(df, "region", "returned")

top_corr_feature    = correlations.index[0]
top_corr_value      = correlations.iloc[0]
highest_risk_region = group_means.index[0]
highest_risk_rate   = group_means.iloc[0]
largest_diff_feat   = summary['mean_diff'].abs().idxmax()

print("=" * 60)
print("  REZUMATUL ANALIZEI RELAȚIILOR DINTRE CARACTERISTICI ȘI ȚINTĂ")
print("=" * 60)
print(f"\n  Set de date: {len(df)} comenzi, {int(df['returned'].sum())} returnate "
      f"({df['returned'].mean() * 100:.1f}% rată de retur)")
print(f"\n  Caracteristica cu cea mai mare corelație : '{top_corr_feature}' (r = {top_corr_value:+.4f})")
print(f"  Cel mai mare decalaj între mediile claselor : '{largest_diff_feat}'")
print(f"    mean_diff = {summary.loc[largest_diff_feat, 'mean_diff']:.2f}")
print(f"\n  Regiunea cu cel mai mare risc    : '{highest_risk_region}' "
      f"(rata de retur = {highest_risk_rate:.1%})")
print(f"\n  Toate corelațiile caracteristicilor cu 'returned':")
for feat, corr in correlations.items():
    bar  = chr(9608) * max(1, int(abs(corr) * 80))
    sign = '+' if corr >= 0 else '-'
    print(f"    {feat:36s} {sign}{abs(corr):.4f}  {bar}")
print("\n  Pași următori recomandați:")
print("  1. Prioritizează caracteristicile cu |r| > 0.05 pentru modelare.")
print("  2. Include 'region' ca variabilă categorială (diferențe vizibile de rată).")
print("  3. Ia în calcul termeni de interacțiune între caracteristicile cu corelație mare.")
print("=" * 60)
print("\n\033[92m  Felicitări pentru finalizarea temei de Cartografiere a Relațiilor dintre\n"
      "  Caracteristici și Țintă! Ai construit un pipeline EDA complet.\033[0m")
```
