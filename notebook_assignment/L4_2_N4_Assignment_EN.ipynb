# Programming Assignment: Bivariate Relationships

## Welcome

**Bivariate analysis** is the process of examining how two variables relate to each other. In machine learning, this is a critical step in exploratory data analysis: it reveals which features are associated with the target, whether numeric relationships are linear or non-linear, and how distributions shift across categories.

### Why this matters

- **Scatter plots** reveal the raw structure of a numeric relationship — are two variables positively correlated, negatively correlated, or independent? Do patterns change when you colour points by the target class?
- **Violin plots** expose *distribution differences* across groups. Unlike a box plot, they show the full shape of the distribution, so you can see whether the two groups have similar spread, skew, or multi-modality.
- **Grouped bar charts** quantify categorical effects: which region has the highest return rate? These charts make decision-relevant differences immediately obvious.
- **Correlation matrices and heatmaps** provide a compact summary of all pairwise linear associations — letting you spot multicollinear features that may destabilise linear models.

> **Tip:** Scatter plots reveal structure; violin plots reveal distribution differences; grouped bar charts reveal categorical effects.

## 7 - Putting It All Together

Once you have implemented all five functions, run the cells below to generate a combined summary dashboard and print key insights from the dataset.

````python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(16, 12))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.45, wspace=0.35)

# Top-left: scatter plot coloured by returned
ax1 = fig.add_subplot(gs[0, 0])
groups_scatter = sorted(df['returned'].unique())
colors_scatter = ['#4C72B0', '#C44E52']
for i, cls in enumerate(groups_scatter):
    mask = df['returned'] == cls
    ax1.scatter(df.loc[mask, 'customer_age'], df.loc[mask, 'order_value'],
                label=f"returned={cls}", color=colors_scatter[i], alpha=0.6,
                edgecolors='white', linewidths=0.3)
ax1.set_title("order_value vs customer_age")
ax1.set_xlabel("customer_age")
ax1.set_ylabel("order_value")
ax1.legend(title='returned')

# Top-right: violin plot of order_value by returned
ax2 = fig.add_subplot(gs[0, 1])
groups_vln = sorted(df['returned'].unique())
data_vln = [df.loc[df['returned'] == g, 'order_value'].dropna().tolist() for g in groups_vln]
ax2.violinplot(data_vln, positions=range(len(groups_vln)), showmedians=True)
ax2.set_xticks(range(len(groups_vln)))
ax2.set_xticklabels([str(g) for g in groups_vln])
ax2.set_title("order_value by returned")
ax2.set_xlabel("returned")
ax2.set_ylabel("order_value")

# Bottom-left: grouped bar rate by region
ax3 = fig.add_subplot(gs[1, 0])
rates = df.groupby('region')['returned'].mean().sort_values(ascending=False)
ax3.bar(rates.index, rates.values, color='#C44E52', edgecolor='white')
for i, (cat, rate) in enumerate(rates.items()):
    ax3.text(i, rate + 0.005, f"{rate:.1%}", ha='center', va='bottom', fontsize=9)
ax3.set_title("returned Rate by region")
ax3.set_xlabel("region")
ax3.set_ylabel("Mean returned")

# Bottom-right: correlation heatmap
ax4 = fig.add_subplot(gs[1, 1])
cols_hm = ['order_value', 'customer_age', 'days_to_ship', 'items_ordered']
corr_hm = df[cols_hm].corr()
im = ax4.imshow(corr_hm.values, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')
fig.colorbar(im, ax=ax4)
ax4.set_xticks(range(len(cols_hm)))
ax4.set_xticklabels(cols_hm, rotation=45, ha='right')
ax4.set_yticks(range(len(cols_hm)))
ax4.set_yticklabels(cols_hm)
for i in range(len(cols_hm)):
    for j in range(len(cols_hm)):
        ax4.text(j, i, f"{corr_hm.iloc[i, j]:.2f}",
                 ha='center', va='center', color='black', fontsize=9)
ax4.set_title("Correlation Heatmap")

fig.suptitle("Bivariate Relationships — Summary Dashboard", fontsize=14, fontweight='bold')
plt.show()
````
