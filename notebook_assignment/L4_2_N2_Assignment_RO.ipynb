## 7 - Punem Totul Cap la Cap

Excelent! Rulează celula de mai jos pentru a vedea toate cele cinci funcții lucrând împreună
pe setul de date cu comenzi.


```python
import matplotlib.pyplot as plt

# ── Afișare alăturată: grafic cu bare sortat + grafic orizontal al proporțiilor ───────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

counts = df['region'].value_counts()
bars = axes[0].bar(counts.index, counts.values, color='#55A868', edgecolor='white')
for bar in bars:
    h = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, h + 1, str(int(h)),
                 ha='center', va='bottom', fontsize=9)
axes[0].set_title("region - Frecvențe")
axes[0].set_xlabel("region")
axes[0].set_ylabel("Număr")

proportions = df['region'].value_counts(normalize=True).sort_values()
axes[1].barh(proportions.index, proportions.values, color='#4C72B0', edgecolor='white')
for i, (cat, prop) in enumerate(proportions.items()):
    axes[1].text(prop + 0.005, i, f"{prop:.1%}", va='center', fontsize=9)
axes[1].set_title("region - Proporțiile Categoriilor")
axes[1].set_xlabel("Proporție")
axes[1].set_ylabel("region")

plt.tight_layout()
plt.show()

# ── Categorii rare la două praguri ─────────────────────────────────────────
print("Categorii rare (threshold=0.12):", identify_rare_categories(df, 'region', threshold=0.12))
print("Categorii rare (threshold=0.20):", identify_rare_categories(df, 'region', threshold=0.20))

# ── Tabel rezumat categorial ───────────────────────────────────────────────
print()
print("Rezumat Categorial:")
print(compute_categorical_summary(df))
```
