# Tema de Programare: Relații Bivariate


## Bine ai venit

**Analiza bivariată** este procesul de examinare a modului în care două variabile se raportează una la cealaltă. În machine learning, acesta este un pas critic în analiza exploratorie a datelor: îți arată ce caracteristici sunt asociate cu ținta, dacă relațiile numerice sunt liniare sau neliniare și cum se schimbă distribuțiile între categorii.

### De ce contează

- **Scatter plot-urile** dezvăluie structura brută a unei relații numerice — două variabile sunt corelate pozitiv, negativ sau independente? Se schimbă tiparele dacă colorezi punctele după clasa țintă?
- **Violin plot-urile** expun *diferențele de distribuție* între grupuri. Spre deosebire de un box plot, ele arată forma completă a distribuției, astfel încât poți vedea dacă cele două grupuri au dispersie, asimetrie sau multimodalitate similare.
- **Graficele cu bare grupate** cuantifică efectele categoriilor: ce regiune are cea mai mare rată de retur? Aceste grafice fac imediat evidente diferențele relevante pentru decizii.
- **Matricele de corelație și heatmap-urile** oferă un rezumat compact al tuturor asocierilor liniare pereche — permițându-ți să observi caracteristicile multicoliniare care pot destabiliza modelele liniare.

> **Pont:** Scatter plot-urile arată structura; violin plot-urile arată diferențele de distribuție; graficele cu bare grupate arată efectele categoriilor.

### Ce vei realiza

În acest notebook vei:
1. Construi un scatter plot pentru două caracteristici numerice, opțional colorat după o țintă binară.
2. Crea violin plot-uri pentru a compara distribuția unei caracteristici între clasele țintei.
3. Calcula și vizualiza ratele de retur pe regiuni folosind grafice cu bare grupate.
4. Calcula o matrice de corelație Pearson pentru un set de coloane numerice.
5. Reprezenta un heatmap de corelație adnotat folosind doar matplotlib.

## 7 - Punem Totul Cap la Cap

După ce implementezi toate cele cinci funcții, rulează celulele de mai jos pentru a genera un dashboard de sinteză și a afișa insight-urile-cheie din setul de date.


```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(16, 12))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.45, wspace=0.35)

# Stânga sus: scatter plot colorat după returned
ax1 = fig.add_subplot(gs[0, 0])
groups_scatter = sorted(df['returned'].unique())
colors_scatter = ['#4C72B0', '#C44E52']
for i, cls in enumerate(groups_scatter):
    mask = df['returned'] == cls
    ax1.scatter(df.loc[mask, 'customer_age'], df.loc[mask, 'order_value'],
                label=f"returned={cls}", color=colors_scatter[i], alpha=0.6,
                edgecolors='white', linewidths=0.3)
ax1.set_title("order_value vs customer_age")
ax1.set_xlabel("customer_age")
ax1.set_ylabel("order_value")
ax1.legend(title='returned')

# Dreapta sus: violin plot pentru order_value după returned
ax2 = fig.add_subplot(gs[0, 1])
groups_vln = sorted(df['returned'].unique())
data_vln = [df.loc[df['returned'] == g, 'order_value'].dropna().tolist() for g in groups_vln]
ax2.violinplot(data_vln, positions=range(len(groups_vln)), showmedians=True)
ax2.set_xticks(range(len(groups_vln)))
ax2.set_xticklabels([str(g) for g in groups_vln])
ax2.set_title("order_value după returned")
ax2.set_xlabel("returned")
ax2.set_ylabel("order_value")

# Stânga jos: rată de retur pe regiune
ax3 = fig.add_subplot(gs[1, 0])
rates = df.groupby('region')['returned'].mean().sort_values(ascending=False)
ax3.bar(rates.index, rates.values, color='#C44E52', edgecolor='white')
for i, (cat, rate) in enumerate(rates.items()):
    ax3.text(i, rate + 0.005, f"{rate:.1%}", ha='center', va='bottom', fontsize=9)
ax3.set_title("Rata lui returned pe region")
ax3.set_xlabel("region")
ax3.set_ylabel("Media lui returned")

# Dreapta jos: heatmap de corelație
ax4 = fig.add_subplot(gs[1, 1])
cols_hm = ['order_value', 'customer_age', 'days_to_ship', 'items_ordered']
corr_hm = df[cols_hm].corr()
im = ax4.imshow(corr_hm.values, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')
fig.colorbar(im, ax=ax4)
ax4.set_xticks(range(len(cols_hm)))
ax4.set_xticklabels(cols_hm, rotation=45, ha='right')
ax4.set_yticks(range(len(cols_hm)))
ax4.set_yticklabels(cols_hm)
for i in range(len(cols_hm)):
    for j in range(len(cols_hm)):
        ax4.text(j, i, f"{corr_hm.iloc[i, j]:.2f}",
                 ha='center', va='center', color='black', fontsize=9)
ax4.set_title("Heatmap de Corelație")

fig.suptitle("Relații Bivariate — Dashboard de Sinteză", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Insight-uri rezumative
numeric_cols = ['order_value', 'days_to_ship', 'items_ordered', 'customer_age', 'days_since_last_purchase']
corr_full = compute_correlation_matrix(df, numeric_cols)

# Cea mai mare corelație absolută în afara diagonalei
corr_vals = corr_full.abs()
np.fill_diagonal(corr_vals.values, 0)
max_pair = corr_vals.stack().idxmax()
max_val = corr_full.loc[max_pair]

# Regiunea cu cea mai mare rată de retur
top_region = df.groupby('region')['returned'].mean().idxmax()
top_rate = df.groupby('region')['returned'].mean().max()

print("=" * 60)
print("  ANALIZĂ BIVARIATĂ — INSIGHT-URI-CHEIE")
print("=" * 60)
print(f"\n  Cea mai puternică corelație pereche:")
print(f"    {max_pair[0]} ↔ {max_pair[1]}  (r = {max_val:.3f})")
print(f"\n  Regiunea cu cea mai mare rată de retur:")
print(f"    {top_region}  ({top_rate:.1%} rată de retur)")
print(f"\n  Insight din scatter:")
print("    Nu există un trend liniar puternic între customer_age și order_value;")
print("    comenzile cu returned=1 sunt distribuite pe toate grupele de vârstă și dimensiunile de comandă.")
print("=" * 60)
```
