Felicitări! Ai stăpânit toolkit-ul de vizualizare statistică din Seaborn. Iată ce ai realizat:

Ideea-cheie: **funcțiile axes-level** îți oferă control precis asupra layout-ului (folosește `ax=ax`), în timp ce **funcțiile figure-level** își gestionează propria pânză și returnează obiecte grid pentru configurarea titlurilor și etichetelor.


```python
# Rezumat final: figură 2×2 care combină toate cele 4 grafice axes-level
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Grafice Statistice cu Seaborn — Rezumat", fontsize=14, fontweight="bold", y=1.01)

# Stânga sus: distribuție
sns.histplot(data=df, x='amount', kde=True, ax=axes[0, 0],
             color='steelblue', edgecolor='white')
axes[0, 0].set_title("Distribuția lui amount")
axes[0, 0].set_xlabel("amount")

# Dreapta sus: violin + swarm
sns.violinplot(data=df, x='category', y='amount', ax=axes[0, 1],
               inner=None, color='lightblue')
sns.swarmplot(data=df, x='category', y='amount', ax=axes[0, 1],
              color='black', size=3, alpha=0.7)
axes[0, 1].set_title("amount după category")

# Stânga jos: heatmap
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, ax=axes[1, 0])
axes[1, 0].set_title("Heatmap de Corelație")

# Dreapta jos: scatter units vs amount pe regiuni
sns.scatterplot(data=df, x='units', y='amount', hue='region',
                ax=axes[1, 1], alpha=0.7)
axes[1, 1].set_title("Units vs Amount pe Regiune")

plt.tight_layout()
plt.show()
```
