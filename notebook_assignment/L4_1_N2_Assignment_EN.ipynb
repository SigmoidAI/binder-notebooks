# Programming Assignment: Seaborn Statistical Plots


Welcome to the **Seaborn Statistical Plots** assignment!

Seaborn builds on Matplotlib and offers two levels of API:

| API Level | Function Type | Returns | Example |
|-----------|--------------|---------|--------|
| **Axes-level** | Draw onto an existing `Axes` | `matplotlib.axes.Axes` | `histplot`, `violinplot`, `heatmap` |
| **Figure-level** | Create their own `Figure` | Grid object (`PairGrid` / `FacetGrid`) | `pairplot`, `lmplot` |

> **Tip:** Always pass an explicit `ax=ax` to axes-level functions to control placement.

**What You Will Do in This Assignment:**

* Plot distributions with histograms and KDE overlays using `histplot`
* Compare categories using violin + swarm overlay plots
* Explore pairwise relationships with `pairplot`
* Visualise correlation matrices with annotated heatmaps
* Fit and display regression lines grouped by category using `lmplot`

```python
# Final summary: 2Ã—2 figure combining all 4 axes-level plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Seaborn Statistical Plots â€” Summary", fontsize=14, fontweight="bold", y=1.01)

# Top-left: distribution
sns.histplot(data=df, x='amount', kde=True, ax=axes[0, 0],
             color='steelblue', edgecolor='white')
axes[0, 0].set_title("amount Distribution")
axes[0, 0].set_xlabel("amount")

# Top-right: violin + swarm
sns.violinplot(data=df, x='category', y='amount', ax=axes[0, 1],
               inner=None, color='lightblue')
sns.swarmplot(data=df, x='category', y='amount', ax=axes[0, 1],
              color='black', size=3, alpha=0.7)
axes[0, 1].set_title("amount by category")

# Bottom-left: heatmap
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, ax=axes[1, 0])
axes[1, 0].set_title("Correlation Heatmap")

# Bottom-right: scatter units vs amount by region
sns.scatterplot(data=df, x='units', y='amount', hue='region',
                ax=axes[1, 1], alpha=0.7)
axes[1, 1].set_title("Units vs Amount by Region")

plt.tight_layout()
plt.show()
```
