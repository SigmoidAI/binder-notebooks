# Programming Assignment: Target Variable Deep Dive

## Welcome

In supervised machine learning, the **target variable** is the column you are trying to predict. Before training any model, you must understand its structure deeply — because flaws in the target corrupt your entire pipeline.

### Why this matters

**Class imbalance** is one of the most common pitfalls in binary classification. When the minority class represents only 18 % of observations (as in our `returned` column), a naive model can achieve 82 % accuracy by *always predicting the majority class* — while being completely useless at detecting returns.

> **Tip:** For imbalanced targets, always use precision / recall / F1, never accuracy alone.

**Target anomalies** are equally dangerous. Unexpected values (e.g., `2` in a binary column) or null entries in the target column will silently distort your model during training if you do not catch them beforehand.

### What you will accomplish

In this notebook you will:
1. Compute a class-balance table and quantify the imbalance.
2. Classify the severity of the imbalance using a four-level scale.
3. Build a publication-quality bar chart of the target distribution.
4. Compute per-class feature means to reveal feature–target associations.
5. Build a robust anomaly detector for target columns.

**Dataset:** 300 simulated e-commerce orders. Target: `returned` (0 = kept, 1 = returned).

## 7 - Putting It All Together

Now we combine all five functions into a complete target variable analysis report.

````python
# Full target analysis pipeline
TARGET = 'returned'
NUMERIC_COLS = ['order_value', 'customer_age', 'days_to_ship', 'items_ordered', 'days_since_last_purchase']

balance = get_class_balance(df, TARGET)
minority_prop = balance['proportion'].min()
severity = classify_imbalance(minority_prop)
stats = compute_target_stats(df, TARGET, NUMERIC_COLS)
anomaly_report = detect_target_anomalies(df, TARGET)

print("=" * 60)
print(f"  TARGET VARIABLE ANALYSIS  —  column: '{TARGET}'")
print("=" * 60)
print("\n[ Class Balance ]")
print(balance.to_string())
print(f"\n[ Imbalance Severity ]")
helper_utils.print_imbalance_report(minority_prop)
print(f"\n[ Anomaly Check ]")
for k, v in anomaly_report.items():
    print(f"  {k}: {v}")
print(f"\n[ Per-Class Feature Means ]")
print(stats.to_string())
````

````python
# Side-by-side visualisation
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(14, 5))

# Left: class distribution bar chart
counts = df[TARGET].value_counts().sort_index()
proportions = counts / len(df)
labels = [str(v) for v in counts.index]
bars = ax_left.bar(labels, counts.values, color=['#4C72B0', '#C44E52'], edgecolor='white')
for bar, prop in zip(bars, proportions.values):
    h = bar.get_height()
    ax_left.text(
        bar.get_x() + bar.get_width() / 2, h + 1,
        f"{prop:.1%}", ha='center', va='bottom', fontsize=11, fontweight='bold',
    )
ax_left.set_title(f"{TARGET} — Class Distribution", fontsize=13)
ax_left.set_xlabel(TARGET)
ax_left.set_ylabel("Count")

# Right: horizontal bar — mean order_value and mean customer_age per class
feature_stats = compute_target_stats(df, TARGET, ['order_value', 'customer_age'])
x = np.arange(len(feature_stats.columns))
width = 0.3
colors_cls = ['#4C72B0', '#C44E52']
for i, (cls_val, row) in enumerate(feature_stats.iterrows()):
    ax_right.barh(
        x + i * width, row.values, height=width,
        label=f"{TARGET}={cls_val}", color=colors_cls[i], edgecolor='white',
    )
ax_right.set_yticks(x + width / 2)
ax_right.set_yticklabels(feature_stats.columns)
ax_right.set_xlabel("Mean value")
ax_right.set_title("Mean Feature Values by Target Class", fontsize=13)
ax_right.legend()

plt.tight_layout()
plt.show()
````
