Welcome to the **Imputation Strategy Comparison** assignment!

Once the missingness mechanism has been characterized, the next step is selecting and applying an appropriate imputation strategy. The theory covers a spectrum of options:

| Strategy | When to use |
|----------|-------------|
| Listwise deletion | Data is MCAR, missingness < 5% |
| Mean imputation | Symmetric, roughly normal distribution, no outliers |
| Median imputation | Skewed distribution or columns with outliers |
| Mode imputation | Categorical / ordinal columns |
| KNN imputation | MAR, multivariate relationships present |
| Iterative (MICE) | MAR, complex multivariate structure |

<a name='4---advanced-imputation'></a>
## 4 - Advanced Imputation

### KNN Imputation

From the theory — KNN imputation finds the $k$ most similar *complete* rows (neighbors) for each row containing a missing value, then fills the gap using the weighted mean of those neighbors' values for that column. It preserves local data structure better than column-wise simple imputation.

**Key trade-off:** Computationally expensive ($O(n^2)$ in naive implementations) and sensitive to feature scale — standardization is required.

### Iterative Imputation (MICE)

MICE treats each column with missing values as a regression target. Each column is iteratively imputed using all other columns as predictors until convergence. This exploits the full multivariate structure of the dataset. Effective under MAR.
