# Programming Assignment: Distribution Analysis


Welcome to the **Distribution Analysis** assignment!

Understanding the shape of your data is one of the first—and most important—steps
in any machine learning or data science project. Distribution analysis answers
fundamental questions such as:

| Question | Tool |
|----------|------|
| What is the typical value? | Mean / Median |
| How spread out is the data? | Std / IQR |
| Is the distribution symmetric? | Skewness |
| Are there unusually heavy tails? | Kurtosis |
| Where are the outliers? | Box plot |

> **Tip:** Always look at both the **histogram** (shape) AND the **box plot** (outliers).
> They reveal different things — a histogram can hide extreme values that the box plot
> clearly flags as outliers.

**What You Will Do in This Assignment:**

* Compute a full set of descriptive statistics for individual columns
* Classify distributions by their skewness and kurtosis profile
* Build a 3-panel visualisation: histogram, KDE, and box plot
* Summarise dispersion statistics across all numeric columns at once
* Automatically flag columns whose skewness exceeds a configurable threshold

## 7 - Putting It All Together

Now we combine all five functions into a complete distribution analysis dashboard.
For each numeric column we compute statistics, classify the distribution shape,
and generate the 3-panel visualisation.


```python
# Full distribution analysis dashboard
import matplotlib
matplotlib.use('Agg')  # suppress interactive display during bulk generation

numeric_cols = df.select_dtypes(include='number').columns.tolist()

print(f"{'Column':<32} {'Mean':>8} {'Std':>8} {'IQR':>8} {'Skewness':>10} {'Shape'}")
print("-" * 80)

for col in numeric_cols:
    stats   = compute_distribution_stats(df, col)
    shape   = classify_distribution(stats['skewness'], stats['kurtosis'])
    print(
        f"{col:<32} {stats['mean']:>8.2f} {stats['std']:>8.2f} "
        f"{stats['iqr']:>8.2f} {stats['skewness']:>10.3f}  {shape}"
    )

print("\nGenerating visualisations...")
for col in numeric_cols:
    fig, _ = plot_distribution_triple(df, col)
    plt.show()

skewed = detect_skewed_columns(df, threshold=1.0)
print(f"\nColumns recommended for log-transform (|skew| >= 1): {skewed}")
```
