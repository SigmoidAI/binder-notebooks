## 1 - Introducere in seriile temporale

**Seria temporala** este o succesiune de puncte de date colectate la intervale succesive de timp. Spre deosebire de datele sectionale, datele de tip serie temporala au o ordine temporala care trebuie respectata.

### Componente-cheie:

1. **Trend (T)**: Crestere sau scadere pe termen lung in date
2. **Sezonalitate (S)**: Tipare regulate care se repeta la intervale fixe
3. **Ciclicitate (C)**: Fluctuatii care apar la intervale neregulate
4. **Reziduu/Zgomot (R)**: Variatia aleatoare ramasa dupa eliminarea celorlalte componente

### Modele de descompunere:

**Model aditiv** (variatie sezoniera constanta):
$$Y_t = T_t + S_t + R_t$$

**Model multiplicativ** (variatie sezoniera proportionala cu nivelul):
$$Y_t = T_t \times S_t \times R_t$$

### Stationaritate:

O serie temporala este **stationara** daca proprietatile sale statistice nu se schimba in timp:
- Media constanta: $E[Y_t] = \mu$
- Varianta constanta: $Var[Y_t] = \sigma^2$
- Autocovarianta constanta: $Cov[Y_t, Y_{t+k}]$ depinde doar de lagul $k$

### Aplicatii:

- **Finante**: preturi bursiere, volume de tranzactionare, analiza riscului
- **Economie**: PIB, inflatie, rata somajului
- **Meteo**: temperatura, precipitatii, tipare ale vantului
- **Sanatate**: semne vitale, raspandirea bolilor, internari in spital
- **Business**: prognoza vanzarilor, managementul stocurilor, planificarea cererii

````python
def plot_time_series(data, title="Time Series", xlabel="Time", ylabel="Value"):
    """Plot time series data."""
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data.values, linewidth=2)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_components(series, trend=None, seasonal=None, residual=None):
    """Plot original series and its components."""
    fig, axes = plt.subplots(4, 1, figsize=(12, 10))
    
    # Original
    axes[0].plot(series.index, series.values, color='blue', linewidth=2)
    axes[0].set_ylabel('Original', fontsize=11)
    axes[0].set_title('Time Series Decomposition', fontsize=14, fontweight='bold')
    axes[0].grid(True, alpha=0.3)
    
    # Trend
    if trend is not None:
        axes[1].plot(trend.index, trend.values, color='red', linewidth=2)
        axes[1].set_ylabel('Trend', fontsize=11)
        axes[1].grid(True, alpha=0.3)
    
    # Seasonal
    if seasonal is not None:
        axes[2].plot(seasonal.index, seasonal.values, color='green', linewidth=2)
        axes[2].set_ylabel('Seasonal', fontsize=11)
        axes[2].grid(True, alpha=0.3)
    
    # Residual
    if residual is not None:
        axes[3].plot(residual.index, residual.values, color='orange', linewidth=1)
        axes[3].set_ylabel('Residual', fontsize=11)
        axes[3].set_xlabel('Time', fontsize=11)
        axes[3].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()


def plot_forecast(train, test, forecast, title="Forecast vs Actual"):
    """Plot training data, test data, and forecasts."""
    plt.figure(figsize=(12, 6))
    
    # Plot training data
    plt.plot(train.index, train.values, label='Training Data', 
             color='blue', linewidth=2)
    
    # Plot test data
    plt.plot(test.index, test.values, label='Actual', 
             color='green', linewidth=2, marker='o', markersize=4)
    
    # Plot forecast
    plt.plot(forecast.index, forecast.values, label='Forecast', 
             color='red', linewidth=2, linestyle='--', marker='s', markersize=4)
    
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel('Time', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.legend(loc='best', fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def calculate_metrics(actual, predicted):
    """Calculate forecasting metrics."""
    mae = np.mean(np.abs(actual - predicted))
    mse = np.mean((actual - predicted) ** 2)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    
    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape
    }

print("Helper functions defined!")
````

## 3 - Descompunerea seriilor temporale

**Descompunerea** separa o serie temporala in componentele sale constitutive pentru a intelege mai bine tiparele de baza.

### Metode de descompunere:

1. **Descompunere clasica**:
   - Foloseste medii mobile pentru a estima trendul
   - Face media valorilor detrendate pentru a estima sezonalitatea
   - Este simpla, dar presupune un tipar sezonier constant

2. **Descompunere STL** (Seasonal and Trend decomposition using Loess):
   - Este mai flexibila si mai robusta
   - Gestioneaza tipare sezoniere care se schimba
   - Este mai putin sensibila la outlieri

### Aditiv vs. multiplicativ:

**Aditiv**: $Y_t = T_t + S_t + R_t$
- Foloseste-l cand variatia sezoniera este constanta
- Este potrivit cand amplitudinea seriei nu se schimba mult

**Multiplicativ**: $Y_t = T_t \times S_t \times R_t$
- Foloseste-l cand variatia sezoniera creste odata cu trendul
- Poate fi convertit in model aditiv folosind log: $\log(Y_t) = \log(T_t) + \log(S_t) + \log(R_t)$

### Aplicatii:
- Identificarea si eliminarea sezonalitatii pentru o modelare mai buna
- Detectarea anomaliilor in reziduuri
- Prognozarea separata a trendului si sezonalitatii

## 4 - Testarea stationaritatii

**Stationaritatea** este un concept crucial in analiza seriilor temporale. Multe metode de forecasting presupun sau necesita date stationare.

### De ce conteaza stationaritatea:

- Face seriile temporale mai usor de modelat
- Proprietatile statistice raman consistente in timp
- Imbunatateste fiabilitatea prognozei
- Este necesara pentru modelele ARIMA

### Testul Augmented Dickey-Fuller (ADF):

Testeaza ipoteza nula: "Seria are radacina unitara (este nestationara)"

**Statistica testului**: Cu cat este mai negativa, cu atat seria este mai probabil stationara

**Regula de decizie**:
- Daca p-value < 0.05: respingi ipoteza nula → **Stationara**
- Daca p-value ≥ 0.05: nu respingi ipoteza nula → **Nestationara**

### Cum faci seria stationara:

1. **Diferentiere**: $Y'_t = Y_t - Y_{t-1}$
   - Elimina trendul
   - Poate fi aplicata de mai multe ori
   - De ordinul intai: $d=1$, de ordinul al doilea: $d=2$

2. **Transformare log**: $Y'_t = \log(Y_t)$
   - Stabilizeaza varianta
   - Converteste multiplicativul in aditiv

3. **Detrendare**: $Y'_t = Y_t - T_t$
   - Elimina componenta de trend
   - Detrendare liniara sau polinomiala

4. **Diferentiere sezoniera**: $Y'_t = Y_t - Y_{t-s}$
   - Elimina sezonalitatea
   - $s$ = perioada sezoniera

## 5 - Analiza autocorelatiei

**Autocorelatia** masoara corelatia dintre o serie temporala si versiunile ei decalate.

### Functia de autocorelatie (ACF):

Corelatia dintre $Y_t$ si $Y_{t-k}$ la lagul $k$:
$$\rho_k = \frac{Cov(Y_t, Y_{t-k})}{\sigma^2}$$

**Proprietati**:
- $\rho_0 = 1$ (corelatie perfecta cu ea insasi)
- $-1 \leq \rho_k \leq 1$
- Scade lent pentru serii nestationare
- Arata efecte directe si indirecte

### Functia de autocorelatie partiala (PACF):

Corelatia dintre $Y_t$ si $Y_{t-k}$ dupa eliminarea efectelor lagurilor intermediare:

**Proprietati**:
- Arata doar relatia directa
- Elimina corelatiile indirecte
- Ajuta la identificarea ordinului AR

### Identificarea modelului:

| Model | Tipar ACF | Tipar PACF |
|-------|-------------|-------------|
| **AR(p)** | Scade treptat | Se opreste dupa lagul p |
| **MA(q)** | Se opreste dupa lagul q | Scade treptat |
| **ARMA(p,q)** | Scade treptat | Scade treptat |

### Intervale de incredere:

Limite de incredere de 95%: $\pm \frac{1.96}{\sqrt{n}}$

Valorile din afara acestor limite sunt semnificative.

## 6 - Metode simple de forecasting

Inainte sa treci la modele complexe, metodele simple de forecasting ofera adesea baseline-uri bune.

### 1. Moving Average (MA):

Media ultimelor $k$ observatii:
$$\hat{Y}_{t+1} = \frac{1}{k} \sum_{i=0}^{k-1} Y_{t-i}$$

**Proprietati**:
- Netezeste fluctuatiile pe termen scurt
- Ramane in urma trendului
- Atribuie ponderi egale tuturor observatiilor din fereastra

### 2. Weighted Moving Average (WMA):

Medie ponderata in care valorile mai recente primesc ponderi mai mari:
$$\hat{Y}_{t+1} = \sum_{i=0}^{k-1} w_i \cdot Y_{t-i}$$

unde $\sum w_i = 1$ si, de obicei, $w_0 > w_1 > ... > w_{k-1}$

### 3. Exponential Smoothing (simplu):

Medie ponderata cu ponderi care scad exponential:
$$\hat{Y}_{t+1} = \alpha Y_t + (1-\alpha) \hat{Y}_t$$

Sau, echivalent:
$$\hat{Y}_{t+1} = \alpha Y_t + \alpha(1-\alpha) Y_{t-1} + \alpha(1-\alpha)^2 Y_{t-2} + ...$$

**Parametri**:
- $\alpha \in [0, 1]$: parametrul de netezire
- $\alpha$ mare (de ex. 0.8): raspunde rapid la schimbari recente
- $\alpha$ mic (de ex. 0.2): netezire mai mare, raspuns mai lent

**Avantaje**:
Usor de implementat
Eficient computational
Are nevoie sa stocheze doar prognoza anterioara
Se adapteaza la schimbari in date

### 4. Double Exponential Smoothing (Holt):

Extinde netezirea simpla pentru a trata trendurile:
- **Nivel**: $L_t = \alpha Y_t + (1-\alpha)(L_{t-1} + T_{t-1})$
- **Trend**: $T_t = \beta(L_t - L_{t-1}) + (1-\beta)T_{t-1}$
- **Prognoza**: $\hat{Y}_{t+h} = L_t + h \cdot T_t$

### Abordare corecta: impartire temporala

Imparte intotdeauna cronologic pentru a respecta ordinea in timp:

```
[----------Date de antrenare----------][--Date de test--]
              Trecut                         Viitor
```

### Metode de impartire:

1. **Holdout simplu**:
   - O singura impartire train/test
   - Antrenezi pe trecut, testezi pe date mai recente
   - Cea mai comuna abordare

2. **Fereastra glisanta**:
   - Fereastra de antrenare de dimensiune fixa
   - Se deplaseaza in timp
   - Utila cand datele recente sunt cele mai relevante

3. **Fereastra in expansiune**:
   - Set de antrenare care creste
   - Testezi pe perioada urmatoare
   - Mai multe date = mai bine pentru termen lung

4. **Cross-validare pentru serii temporale**:
   ```
   Fold 1: [train] [test]
   Fold 2: [train---] [test]
   Fold 3: [train------] [test]
   ```

### Bune practici:

Foloseste 70-80% pentru antrenare, 20-30% pentru testare
Ia in calcul un gap intre train si test (orizontul de forecasting)
Valideaza pe mai multe perioade, daca este posibil
Fii atent la tiparele sezoniere
Nu amesteca niciodata datele de tip serie temporala
Nu folosi date din viitor pentru a prezice trecutul

**Pentru Moving Average simplu:**
* Parcurge seria de la `window` pana la final
* Pentru fiecare pozitie: `ma[i] = np.mean(series[i-window:i])`
* Primele valori `(window-1)` vor fi NaN

**Pentru Moving Average ponderat:**
* Creeaza ponderi: `weights = np.arange(1, window+1)` (liniar)
* Normalizeaza: `weights = weights / weights.sum()`
* Calculeaza: `wma[i] = np.dot(series[i-window:i], weights)`

**Ponderi alternative:**
* Exponentiale: `np.exp(np.arange(window))`
* Personalizate: orice secventa cu suma 1

</details>

### Exercitiul 8 - Exponential Smoothing de la zero

**Sarcina ta:**

Implementeaza single exponential smoothing de la zero folosind formula recursiva:
$$S_t = \alpha Y_t + (1-\alpha) S_{t-1}$$

unde $S_0 = Y_0$ (prima observatie)

<details>
  <summary><b><font color="green">Indicii de cod (Click pentru a extinde daca ai nevoie de ajutor)</font></b></summary>
  
**Pentru initializare:**
* Prima valoare netezita: `smoothed[0] = series[0]`

**Pentru recursie:**
* Parcurge de la 1 la n
* Actualizeaza: `smoothed[t] = alpha * series[t] + (1 - alpha) * smoothed[t-1]`

**Pentru forecasting:**
* Un pas inainte: foloseste ultima valoare netezita
* Mai multi pasi: repeta aceeasi valoare (forecast de nivel)

**Interpretarea lui alpha:**
* Alpha mare (0.8): raspunde rapid, netezeste mai putin
* Alpha mic (0.2): netezeste mai mult, raspunde mai lent

</details>


````python
# GRADED FUNCTION: exponential_smoothing_scratch
def exponential_smoothing_scratch(series, alpha=0.3):
    """
    Compute single exponential smoothing from scratch.
    
    Args:
        series: Time series data (array-like)
        alpha: Smoothing parameter (0 to 1)
    
    Returns:
        Smoothed values (numpy array, same length as input)
    """
    series = np.array(series)
    n = len(series)
    smoothed = np.zeros(n)
    
    ### ÎNCEPUT DE COD AICI ### (≈ 6 lines)
    # Initialize with first observation
    smoothed[0] = None
    
    # Apply exponential smoothing recursively
    for t in range(1, n):
        # S_t = alpha * Y_t + (1 - alpha) * S_{t-1}
        smoothed[t] = None
    ### SFÂRȘIT DE COD AICI ###
    
    return smoothed
````
