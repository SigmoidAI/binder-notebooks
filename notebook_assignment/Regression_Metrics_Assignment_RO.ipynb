## 1 - Introducere in metricile de regresie

**Metricile de regresie** cuantifica cat de bine se potrivesc valorile continue prezise cu valorile reale. Spre deosebire de metricile de clasificare, care masoara corectitudinea predictiilor discrete, metricile de regresie masoara magnitudinea erorilor de predictie.

### Concepte de baza:

**Reziduuri**: diferenta dintre valorile reale si cele prezise
$$e_i = y_i - \hat{y}_i$$

**Mean Squared Error (MSE)**: media erorilor la patrat
$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

**Root Mean Squared Error (RMSE)**: radacina patrata a MSE
$$\text{RMSE} = \sqrt{\text{MSE}} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

**Mean Absolute Error (MAE)**: media erorilor absolute
$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

**R-Squared (R²)**: proportia de varianta explicata
$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2} = 1 - \frac{SS_{res}}{SS_{tot}}$$

### Diferente cheie:

**MSE vs MAE**:
- MSE penalizeaza mai puternic erorile mari (termenul patrat)
- MAE trateaza toate erorile la fel
- Unitatile MSE: unitati la patrat ale tintei
- Unitatile MAE: aceleasi unitati ca tinta

**RMSE vs MAE**:
- RMSE este in aceleasi unitati ca tinta (ca MAE)
- RMSE este mai sensibil la outlieri
- MAE este mai robust la outlieri

**Interpretarea R²**:
- R² = 1: predictii perfecte
- R² = 0: modelul performeaza ca un baseline cu media
- R² < 0: modelul este mai slab decat baseline-ul cu media

### Selectarea metricii:

- **Exista outlieri**: foloseste MAE (robust)
- **Erorile mari sunt foarte costisitoare**: foloseste MSE sau RMSE
- **Ai nevoie de interpretabilitate**: foloseste MAE sau RMSE (aceleasi unitati)
- **Compari modele**: foloseste R² sau Adjusted R²
- **Costuri specifice business-ului**: foloseste metrici personalizate

### Functii ajutatoare


````python
def plot_predictions(y_true, y_pred, title="Predicted vs Actual"):
    """Plot predicted vs actual values with perfect prediction line."""
    plt.figure(figsize=(10, 6))
    plt.scatter(y_true, y_pred, alpha=0.5, edgecolors='k', linewidth=0.5)
    
    # Perfect prediction line
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
    
    plt.xlabel('Actual Values', fontsize=12)
    plt.ylabel('Predicted Values', fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_residuals(y_true, y_pred, title="Residual Plot"):
    """Plot residuals vs predicted values."""
    residuals = y_true - y_pred
    
    plt.figure(figsize=(10, 6))
    plt.scatter(y_pred, residuals, alpha=0.5, edgecolors='k', linewidth=0.5)
    plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
    plt.xlabel('Predicted Values', fontsize=12)
    plt.ylabel('Residuals', fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_residual_distribution(y_true, y_pred, title="Residual Distribution"):
    """Plot histogram and Q-Q plot of residuals."""
    residuals = y_true - y_pred
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Histogram
    axes[0].hist(residuals, bins=50, edgecolor='k', alpha=0.7)
    axes[0].axvline(x=0, color='r', linestyle='--', linewidth=2)
    axes[0].set_xlabel('Residuals', fontsize=12)
    axes[0].set_ylabel('Frequency', fontsize=12)
    axes[0].set_title('Residual Histogram', fontsize=14, fontweight='bold')
    axes[0].grid(True, alpha=0.3)
    
    # Q-Q plot
    stats.probplot(residuals, dist="norm", plot=axes[1])
    axes[1].set_title('Q-Q Plot', fontsize=14, fontweight='bold')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()


def plot_metric_comparison(models_dict, metrics_dict):
    """Plot comparison of multiple models across multiple metrics."""
    df = pd.DataFrame(metrics_dict).T
    
    fig, axes = plt.subplots(1, len(df.columns), figsize=(5*len(df.columns), 5))
    if len(df.columns) == 1:
        axes = [axes]
    
    for idx, metric in enumerate(df.columns):
        df[metric].plot(kind='bar', ax=axes[idx], color='skyblue', edgecolor='k')
        axes[idx].set_title(metric, fontsize=14, fontweight='bold')
        axes[idx].set_ylabel('Score', fontsize=12)
        axes[idx].set_xlabel('Model', fontsize=12)
        axes[idx].grid(True, alpha=0.3, axis='y')
        axes[idx].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()


print("Helper functions defined!")
````

## 2 - Metrici bazate pe eroare

**Metricile bazate pe eroare** masoara abaterea dintre predictii si valorile reale. Acestea sunt cele mai fundamentale metrici de regresie.

### Mean Squared Error (MSE):

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

**Proprietati:**
- Mereu nenegativa (≥ 0)
- Penalizeaza puternic erorile mari (cvadratic)
- Unitati: unitatile la patrat ale variabilei tinta
- Diferentiabila (utila pentru optimizare bazata pe gradient)
- Sensibila la outlieri

### Root Mean Squared Error (RMSE):

$$\text{RMSE} = \sqrt{\text{MSE}}$$

**Proprietati:**
- Are aceleasi unitati ca variabila tinta (usor de interpretat)
- Continua sa penalizeze erorile mari mai mult decat MAE
- Reprezinta abaterea standard a reziduurilor
- Sensibila la outlieri

### Mean Absolute Error (MAE):

$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

**Proprietati:**
- Aceleasi unitati ca variabila tinta
- Penalizare liniara a erorilor
- Robust la outlieri
- Nu este diferentiabila in zero (provocare pentru optimizare)

### Cand folosesti fiecare:

**Foloseste MSE/RMSE cand:**
- Erorile mari sunt disproportionat de costisitoare
- Antrenezi retele neuronale (functie de pierdere diferentiabila)
- Te astepti la o distributie normala a erorilor

**Foloseste MAE cand:**
- Toate erorile sunt la fel de importante
- Exista outlieri in date
- Ai nevoie de o metrica robusta
- Interpretabilitatea este importanta

### Relatia dintre ele:

Pentru orice set de date: $\text{MAE} \leq \text{RMSE}$

O diferenta mare indica prezenta outlierilor sau o varianta mare a erorilor.

## 3 - R-Squared si Adjusted R-Squared

**R-squared (R²)** masoara proportia din varianta variabilei tinta care este explicata de model.

### R-Squared (coeficientul de determinare):

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

unde:
- $SS_{res}$ = suma patratelor reziduurilor (eroarea modelului)
- $SS_{tot}$ = suma totala a patratelor (varianta din date)
- $\bar{y}$ = media valorilor reale

**Interpretare:**
- R² = 1.0: predictii perfecte (toata varianta este explicata)
- R² = 0.0: modelul este la fel de bun ca predictia cu media
- R² < 0.0: modelul este mai slab decat media (potrivire slaba)
- R² = 0.7: modelul explica 70% din varianta

### Adjusted R-Squared:

$$R^2_{adj} = 1 - \frac{(1-R^2)(n-1)}{n-p-1}$$

unde:
- $n$ = numarul de esantioane
- $p$ = numarul de caracteristici

**Scop:**
- Penalizeaza adaugarea de caracteristici care nu imbunatatesc modelul
- Tine cont de complexitatea modelului
- Poate scadea cand adaugi caracteristici irelevante
- Este mai potrivit pentru compararea modelelor cu numere diferite de caracteristici

### Limitele lui R²:

1. **Creste mereu cand adaugi mai multe caracteristici** (chiar si aleatoare)
2. **Nu este comparabil intre seturi de date diferite**
3. **Nu indica daca modelul este potrivit**
4. **Poate fi inselator in relatii neliniare**
5. **Este sensibil la outlieri**

### Valori bune pentru R²:

Depind de context:
- **Fizica/Inginerie**: de obicei se asteapta R² > 0.9
- **Economie/Stiinte sociale**: adesea R² > 0.5 este bun
- **Biologie/Comportament**: R² > 0.3 poate fi acceptabil

**Verifica mereu:**
- Graficele reziduurilor (mai importante decat R²)
- R² pe train vs test (pentru overfitting)
- Cunostintele de domeniu si asteptarile

## 4 - Analiza reziduurilor

**Analiza reziduurilor** este esentiala pentru validarea ipotezelor modelului de regresie si diagnosticarea problemelor.

### Ce sunt reziduurile?

$$e_i = y_i - \hat{y}_i$$

Reziduurile sunt erorile dintre valorile reale si cele prezise. Analiza tiparelor lor scoate la iveala probleme ale modelului.

### Ipoteze cheie de verificat:

**1. Liniaritate:**
- Relatia dintre caracteristici si tinta este liniara
- **Verificare**: graficul Reziduuri vs Predictii ar trebui sa arate imprastiere aleatoare
- **Incalcare**: un tipar curbat indica o relatie neliniara

**2. Homoscedasticitate (varianta constanta):**
- Varianta erorii este constanta pentru toate nivelurile predictiei
- **Verificare**: graficul Reziduuri vs Predictii ar trebui sa aiba raspandire constanta
- **Incalcare**: forma de palnie indica heteroscedasticitate

**3. Normalitatea reziduurilor:**
- Reziduurile urmeaza o distributie normala
- **Verificare**: grafic Q-Q, histograma
- **Incalcare**: histograma asimetrica, puncte in afara diagonalei in graficul Q-Q

**4. Independenta:**
- Reziduurile sunt independente (fara autocorelatie)
- **Verificare**: graficul Reziduuri vs ordine/timp
- **Incalcare**: tipare in functie de timp/ordine

### Grafice de diagnostic:

**Reziduuri vs Predictii:**
- Imprastiere aleatoare -> bine
- Tipar curbat -> neliniaritate
- Forma de palnie -> heteroscedasticitate
- Puncte departate de zero -> outlieri

**Grafic Q-Q:**
- Puncte pe diagonala -> distributie normala
- Puncte in afara diagonalei -> nenormalitate
- Cozi grele -> prezenta outlierilor

**Histograma reziduurilor:**
- In forma de clopot, centrata pe zero -> bine
- Asimetrica -> nenormalitate
- Varfuri multiple -> amestec de populatii

### Probleme frecvente:

**Outlieri:**
- Reziduuri cu valori absolute mari
- Pot distorsiona potrivirea modelului
- **Solutie**: investigheaza, elimina daca sunt erori sau foloseste metode robuste

**Puncte cu leverage mare:**
- Valori extreme ale caracteristicilor
- Pot influenta disproportional modelul
- **Verificare**: distanta Cook, grafice de leverage

**Neliniaritate:**
- **Solutie**: transforma caracteristicile, adauga termeni polinomiali sau foloseste un model neliniar

**Heteroscedasticitate:**
- **Solutie**: transforma tinta (log), regresie ponderata, erori standard robuste

## 5 - Compararea metricilor

**Metrici diferite surprind aspecte diferite ale performantei modelului.** Compararea mai multor modele necesita evaluarea lor simultana pe mai multe metrici.

**Limitari ale unei singure metrici:**
- MSE: puternic influentat de outlieri

- RMSE: unitatile pot sa nu fie intuitive

**Ce necesita o evaluare cuprinzatoare:**
- Magnitudinea erorii (MAE, RMSE)
- Sensibilitate la outlieri (diferenta MAE vs RMSE)
- Varianta explicata (R², Adjusted R²)
- Tipare in reziduuri (grafice de diagnostic)

### Cadru de comparare a modelelor:

**1. Antreneaza mai multe modele:**
- Linear Regression (baseline)
- Ridge Regression (regularizare L2)

- MSE, RMSE, MAE
- R², Adjusted R²
- Eroare maxima, eroare mediana

- Discrepante intre metrici (prezenta outlierilor)

- In functie de prioritatile de business
- Tine cont de interpretabilitate
- Valideaza pe un set holdout

### Analiza discrepantelor dintre metrici:

**R² mic dar MAE mic:**
- Predictiile modelului sunt in medie corecte

## 6 - Metrici personalizate

**Problemele specifice de business necesita adesea functii de pierdere personalizate** care se aliniaza cu costurile si prioritatile reale.

### De ce metrici personalizate?

Metricile standard (MSE, MAE, R²) pot sa nu reflecte:
- **Costuri asimetrice**: supraestimare vs subestimare
- **Penalitati neliniare**: anumite intervale de eroare sunt mai critice
- **Constrangeri de business**: cerinte minime de acuratete
- **Nevoi specifice domeniului**: standarde industriale

### Metrici personalizate comune:

**1. Pierdere asimetrica:**
$$L_{asymm} = \frac{1}{n} \sum_{i=1}^{n} \begin{cases} 
\alpha \cdot (y_i - \hat{y}_i)^2 & \text{if } y_i > \hat{y}_i \\
\beta \cdot (y_i - \hat{y}_i)^2 & \text{if } y_i \leq \hat{y}_i
\end{cases}$$

**Caz de utilizare**: prognoza cererii
- Subestimare (lipsa de stoc): cost mare $\alpha$
- Supraestimare (inventar in exces): cost mai mic $\beta$

**2. Quantile Loss (Pinball Loss):**
$$L_q(y, \hat{y}) = \begin{cases}
q \cdot (y - \hat{y}) & \text{if } y \geq \hat{y} \\
(q-1) \cdot (y - \hat{y}) & \text{if } y < \hat{y}
\end{cases}$$

**Caz de utilizare**: regresie pe cuantile
- $q = 0.5$: mediana (robusta la outlieri)
- $q = 0.9$: predictie pentru percentila 90

**3. Pierdere bazata pe procent:**
$$L_{pct} = \frac{1}{n} \sum_{i=1}^{n} \left|\frac{y_i - \hat{y}_i}{y_i}\right| \times 100$$

**Caz de utilizare**: cand eroarea relativa conteaza
- Aceeasi eroare absoluta are impact diferit la scari diferite
- Eroare de $1k la o casa de $10k vs $100k

**4. Accuracy bazat pe prag:**
$$\text{Acc}_{\epsilon} = \frac{1}{n} \sum_{i=1}^{n} \mathbb{1}\{|y_i - \hat{y}_i| \leq \epsilon\}$$

**Caz de utilizare**: succes/esec binar
- Predictiile in interiorul tolerantei sunt "bune"
- Cele din afara tolerantei sunt "rele"

### Proiectarea metricilor personalizate:

**Pasi:**
1. **Identifica costurile de business**: care sunt costurile reale ale erorilor?
2. **Cuantifica asimetria**: supra/subpredictiile sunt la fel de grave?
3. **Defineste pragurile**: ce niveluri de eroare sunt acceptabile?
4. **Implementeaza functia**: codifica pierderea personalizata
5. **Valideaza**: testeaza pe cazuri cunoscute
6. **Optimizeaza**: antreneaza modele care minimizeaza pierderea personalizata

### Exemple de scenarii:

**Imobiliare:**
- Supra-pretuire: casa ramane nevanduta (costuri de mentinere)
- Sub-pretuire: bani lasati pe masa
- Personalizare: pierdere asimetrica cu cunostinte de domeniu

**Cerere de energie:**
- Supra-predictie: cost de productie in exces
- Sub-predictie: pene de curent (catastrofal)
- Personalizare: penalizeaza puternic subpredictia

**Dozare medicala:**
- Subdoza: tratament ineficient
- Supradoza: efecte secundare toxice
- Personalizare: marje de siguranta cu penalizare neliniara

## 7 - Implementare de la zero: toate metricile de regresie

Implementarea tuturor metricilor de regresie de la zero aprofundeaza intelegerea fundamentelor lor matematice si a relatiilor dintre ele.

### Metrici de implementat:

**1. MSE (Mean Squared Error)**
```python
MSE = (1/n) * Σ(y_true - y_pred)²
```

**2. RMSE (Root Mean Squared Error)**
```python
RMSE = √MSE
```

**3. MAE (Mean Absolute Error)**
```python
MAE = (1/n) * Σ|y_true - y_pred|
```

**4. R² (Coefficient of Determination)**
```python
R² = 1 - (SS_res / SS_tot)
SS_res = Σ(y_true - y_pred)²
SS_tot = Σ(y_true - mean(y_true))²
```

**5. Adjusted R²**
```python
Adj_R² = 1 - ((1-R²) * (n-1)) / (n-p-1)
```

### Detalii importante de implementare:

**1. Operatii pe array-uri:**
- Foloseste NumPy pentru calcule eficiente
- Vectorizeaza operatiile (evita buclele)
- Asigura-te ca array-urile au aceeasi forma

**2. Cazuri limita:**
- Impartire la zero in R² (cand SS_tot = 0)
- Array-uri goale
- Array-uri cu o singura valoare

**3. Stabilitate numerica:**
- Gestioneaza valori foarte mici / foarte mari
- Foloseste precizie adecvata

**4. Validare:**
- Testeaza fata de implementari sklearn
- Foloseste cazuri de test cunoscute
- Verifica relatiile matematice

- MAE pentru predictii robuste

**Consum energetic:**
- RMSE pentru planificarea capacitatii
- R² pentru compararea modelelor
- Analiza reziduurilor pentru tipare de consum

**Modelare financiara:**
- Metrici multiple pentru evaluare cuprinzatoare
- Metrici personalizate pentru randamente ajustate la risc
- Accuracy pe prag pentru intervale acceptabile de eroare

### Subiecte avansate:

**Cross-Validation:**
- K-fold cross-validation pentru evaluare robusta
- Time series cross-validation pentru date temporale
- Esantionare stratificata pentru distributii consistente

**Intervale de incredere:**
- Prediction intervals pentru cuantificarea incertitudinii
- Metode bootstrap pentru increderea in metrici
- Abordari bayesiene pentru predictii probabilistice

**Analiza avansata a reziduurilor:**
- Distanta Cook pentru puncte influente
- Grafice de leverage pentru observatii cu leverage mare
- Partial residual plots pentru relatii intre caracteristici
- Testul Durbin-Watson pentru autocorelatie

**Regresie robusta:**
- Huber loss (combina MSE si MAE)
- RANSAC pentru potrivire rezistenta la outlieri
- Regresie pe cuantile pentru predictii robuste

### Checklist de diagnostic:

**Inainte sa ai incredere in model:**
- [ ] Metrici multiple calculate (MSE, MAE, R²)
- [ ] Performanta pe train vs test comparata
- [ ] Graficul reziduurilor nu arata tipare
- [ ] Reziduurile sunt aproximativ normal distribuite
- [ ] Varianta este constanta intre predictii (homoscedasticitate)
- [ ] Outlierii au fost identificati si investigati
- [ ] Au fost evaluate metrici relevante pentru business
- [ ] A fost efectuata cross-validare
- [ ] Modelul are sens intuitiv
