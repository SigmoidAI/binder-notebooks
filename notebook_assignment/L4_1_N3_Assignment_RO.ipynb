# Tema de Programare: Vizualizare Interactivă cu Plotly Express


Bine ai venit la tema despre **Vizualizare Interactivă cu Plotly Express**!

Avantajul-cheie al lui Plotly față de bibliotecile statice este **interactivitatea** — fiecare grafic pe care îl creezi suportă implicit tooltip-uri la hover, zoom, pan și descărcare PNG printr-un singur click. Nu este nevoie de configurare suplimentară.

Plotly expune două niveluri de API:

| Nivel API | Descriere | Returnează |
|-----------|-----------|------------|
| **`plotly.express` (px)** | Grafice de nivel înalt, dintr-o singură linie | `go.Figure` |
| **`plotly.graph_objects` (go)** | Nivel jos, control complet | `go.Figure` |

> **Pont:** Fiecare funcție `px.*` returnează un `go.Figure`. Folosește `fig.update_layout()` și `fig.update_traces()` pentru a rafina orice grafic creat de oricare dintre cele două API-uri.

* Personalizezi layout-urile figurilor cu titluri, etichete de axe și setări de font

<a name='7---putting-it-all-together'></a>
## 7 - Punem Totul Cap la Cap

Acum să combinăm toate cele cinci funcții pentru a construi un mini dashboard de vânzări. Vom afișa trei perspective coordonate asupra aceluiași set de date:

1. **Scatter** — venit vs unități, colorat după categorie
2. **Serii temporale** — evoluția venitului, colorată după categorie
3. **Bare grupate** — venit pe categorie, subgrupat după statusul livrării

Fiecare grafic este personalizat complet, iar vizualizarea scatter finală este exportată ca fișier HTML.


```python
# --- Graficul 1: Scatter ---
scatter = create_scatter(df, x='units', y='amount', color='category',
                         hover_data=['region', 'status'])
scatter = customize_layout(scatter, title='Venit vs Unități (după Categorie)',
                           xaxis_title='Unități Comandate',
                           yaxis_title='Venit (USD)')
scatter.show()

# --- Graficul 2: Serii temporale ---
ts = create_time_series(df, date_col='order_date', value_col='amount',
                        color_col='category')
ts = customize_layout(ts, title='Venit Săptămânal pe Categorie',
                      xaxis_title='Data Comenzii',
                      yaxis_title='Venit (USD)')
ts.show()

# --- Graficul 3: Bare grupate ---
bar = create_grouped_bar(df, x='category', y='amount', color='status')
bar = customize_layout(bar, title='Venit pe Categorie și Status de Livrare',
                       xaxis_title='Categorie',
                       yaxis_title='Venit (USD)')
bar.show()

# --- Exportă graficul scatter ---
save_figure_html(scatter, '/tmp/L4_1_N3_dashboard_scatter.html')
print('Scatter-ul dashboard-ului a fost salvat în /tmp/L4_1_N3_dashboard_scatter.html')
```
