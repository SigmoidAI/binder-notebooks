category, or customer segment. Before modelling, it is essential to understand:

| Question | Technique |

<a name='7---putting-it-all-together'></a>
## 7 - Putting It All Together

Excellent work! Run the cell below to see all five functions working together
on the orders dataset.


```python
import matplotlib.pyplot as plt

# ── Side-by-side: sorted bar chart + horizontal proportion chart ───────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

counts = df['region'].value_counts()
bars = axes[0].bar(counts.index, counts.values, color='#55A868', edgecolor='white')
for bar in bars:
    h = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, h + 1, str(int(h)),
                 ha='center', va='bottom', fontsize=9)
axes[0].set_title("region - Value Counts")
axes[0].set_xlabel("region")
axes[0].set_ylabel("Count")

proportions = df['region'].value_counts(normalize=True).sort_values()
axes[1].barh(proportions.index, proportions.values, color='#4C72B0', edgecolor='white')
for i, (cat, prop) in enumerate(proportions.items()):
    axes[1].text(prop + 0.005, i, f"{prop:.1%}", va='center', fontsize=9)
axes[1].set_title("region - Category Proportions")
axes[1].set_xlabel("Proportion")
axes[1].set_ylabel("region")

plt.tight_layout()
plt.show()

# ── Rare categories at two thresholds ─────────────────────────────────────────
print("Rare categories (threshold=0.12):", identify_rare_categories(df, 'region', threshold=0.12))
print("Rare categories (threshold=0.20):", identify_rare_categories(df, 'region', threshold=0.20))

# ── Categorical summary table ──────────────────────────────────────────────────
print()
print("Categorical Summary:")
print(compute_categorical_summary(df))
```
